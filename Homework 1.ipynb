```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
import matplotlib.pyplot as plt
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/postoperative-patient-data/post-operative.data'
# download the file
raw_data = urlopen(url)
```

```python
dataset = np.genfromtxt(raw_data, delimiter=",", dtype=None, encoding = None)
```

Data was imported as an array of strings. 

```python
dataset.shape
```

```python
dataset
```

```python
patdata = pd.DataFrame(dataset)
```

```python
patdata.columns  = ['L-CORE', 'L-SURF', 'L-O2', 'L-BP', 'SURF-STBL', 'CORE-STBL', 'BP-STBL', 'COMFORT', 'Decision']
```

Dataset array was converted into a pandas dataframe to be able to display and manipulate the data easily. The column titles were added after the dataframe was created. 

```python
patdata.head()
```

```python
patdata['COMFORT'] = patdata['COMFORT'].apply(pd.to_numeric, errors = 'coerce')
```

Comfort column was converted from a column of strings to a numeric (float) column. This allows for the plotting of the different attributes against the Comfort levels of the patients. It also forces the few rows that don't have comfort data to be converted from '?' to 'NaN.' 

```python
patdata
```

```python
patdata.isnull().sum()
```

```python
df = patdata
df = df.dropna()
```

The rows within the dataframe that had null data for the comfort were identified and dropped from the dataset that will be used to graph the different comparisons. These rows were dropped because they did not include the comfort levels of those patients. It would make a big difference to the data set if the points were dropped, and considering that the data is subjective to each patient, imputing those points wouldn't really make sense, especially with such a limited amount of categories in the x-axis. There doesn't seem to be a reliable to way to impute those data points. 

```python
import seaborn as sns
```

```python
sns.swarmplot(x=df['L-CORE'], y=df['COMFORT'])
```

Seaborn was used to plot the data in a categorical scatterplot. The columns from the dataframe were directly loaded into the plat as the x and y axis. Swarm plots were used to represent the distribution of comfort levels amongst the patients. 

This chart shows core body temperature against comfort. The graph shows that the there is no clear correlation between core temperature and comfort level because each of the categorical points for core body termperature have a similar distribution of comfort. Although it can be noted that mid core temp have the most patients that had a high comfort.
